```python
import json 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from PIL import Image
```

```python
path_topk_dinov3 = "/media/jbhayet/Data/datasets/croco-experiments/results/retrieval/top10_dinov3.jsonl"
df = pd.read_json(path_topk_dinov3, lines=True) 
```

```python
idx = 0
q = df["query_path"][idx]
topk = df["topk"][idx]
```

```python
q_img = np.asarray(Image.open(q))
fig = plt.figure(figsize=(5, 5))
plt.imshow(q_img); plt.title("query image")
plt.show()
```

```python
fig, ax = plt.subplots(2, 5, figsize=(15, 6))
axis = ax.flatten()
for i, retrieve in enumerate(topk):
    ax = axis[i]
    rank = retrieve.get("rank")
    score = retrieve.get("score")
    path = retrieve.get("abs_path")
    img = np.asarray(Image.open(path))
    ax.imshow(img); ax.set_title(f"rank: {rank}, score: {score:.4g}")
plt.tight_layout()
plt.show()
```
