```python
import json 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from PIL import Image
```

### map hl query spot 

```python
path_topk_dinov3 = "/home/sergio/vis_loc/data/HYDRO_map_hl_query_spot.json"
df = pd.read_json(path_topk_dinov3) 
df.head()
```

```python
idx = 0
q = df["path"][idx]
topk = df["topk"][idx]
topk
```

```python
q_img = np.asarray(Image.open(q))
fig = plt.figure(figsize=(5, 5))
plt.imshow(q_img); plt.title("query image")
plt.show()
```

```python
fig, ax = plt.subplots(2, 5, figsize=(15, 6))
axis = ax.flatten()
for i, retrieve in enumerate(topk):
    ax = axis[i]
    rank = retrieve.get("rank")
    score = retrieve.get("score")
    path = retrieve.get("path")
    img = np.asarray(Image.open(path))
    ax.imshow(img); ax.set_title(f"rank: {rank}, score: {score:.4g}")
plt.tight_layout()
plt.show()
```

### map hl query ios 

```python
path_topk_dinov3 = "/home/sergio/vis_loc/data/HYDRO_map_hl_query_ios.json"
df = pd.read_json(path_topk_dinov3) 
df.head()
```

```python
idx = 2
q = df["path"][idx]
topk = df["topk"][idx]

q_img = np.asarray(Image.open(q))

fig, ax = plt.subplots(3, 5, figsize=(15, 6))
axis = ax.flatten()
for i in range(len(topk) + 1):
    ax = axis[i]
    if i == 0:
        ax.imshow(q_img); ax.set_title("query image")
        continue 
    retrieve = topk[i-1]
    rank = retrieve.get("rank")
    score = retrieve.get("score")
    path = retrieve.get("path")
    img = np.asarray(Image.open(path))
    ax.imshow(img); ax.set_title(f"rank: {rank}, score: {score:.4g}")
plt.tight_layout()
plt.show()
```
